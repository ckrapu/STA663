```python
import numpy as np
import scipy as sc
import statsmodels.api as stats
from __future__ import division
import string
import nltk
import sklearn
import sys
import copy
# assume format is documents delimited by line 

def text_munge(filename):
    
    f = open(filename)
    text = f.read()
    text = text.translate(string.maketrans("",""), string.punctuation)
    text = text.lower()
    split_text = text.split("\r\n")
    split_text = filter(None,split_text)

    docs=[]
    for paragraph in split_text:
        doc = paragraph.split(" ")
        docs.append(doc)


    uniqueWords=[]
    wordCounts=[]


    for doc in docs:
    
        for word in doc:
       
            if word in uniqueWords:
                for v,term in enumerate(uniqueWords):
                    if term==word:
                        wordCounts[v]+=1
                
            else:
                uniqueWords.append(word)
                wordCounts.append(1)
    
    V=len(uniqueWords) 
    dictionary = dict(zip(uniqueWords,range(0,V)))

    integer_docs=[]

    for doc in docs:
        integer_doc=[]
        for word in doc:
            integer_doc.append(dictionary[word])
        
        integer_docs.append(integer_doc)
 

    dictionary = {v: k for k, v in dictionary.items()}
    
    return dictionary,integer_docs,wordCounts,uniqueWords

def initialize_assignments(V,M,tree,L,corpus,c_ml):
    K=len(tree)
    z = copy.deepcopy(corpus)
    NWZ = np.zeros([V,K])
    NZM = np.zeros([K,M])
    NZ = np.zeros([K])
    for m,document in enumerate(corpus):
    
        for n,word in enumerate(document):
            
            z_out = np.random.multinomial(1,np.repeat(1.0/L,L)).dot(c_ml[m,:])
            NZM[z_out,m]+=1
            NWZ[word,z_out]+=1
            NZ[z_out]+=1
        
            # using list of lists so we have to use list access notation instead of slice
            z[m][n] = z_out
    return z,NWZ,NZM,NZ


def sample_assignment(m,corpus,V,alpha,beta,z,NWZ,NZM,NZ,c_ml):
    document = corpus[m]
    K = len(c_ml[0])
    
        


    # vector of probabilities; each entry indicates the probability that the new label assignment will be k
    p_z= np.zeros([K,])
    for n,word in enumerate(document):
               
            NWZ[word,z[m][n]]-=1
            NZ[z[m][n]]-=1
            NZM[z[m][n],m]-=1
            
            assert NWZ[word,z[m][n]]>=0
            assert NZ[z[m][n]]>=0
            assert NZM[z[m][n],m]>=0
            
            for i in range(0,K):
                p_z[i] = (max(NWZ[word,c_ml[m,i]],0)+beta)/(NWZ[:,c_ml[m,i]].sum()+V*beta)*(max(NZM[c_ml[m,i],m],0)+alpha)
                assert p_z[i]>0
                
            
            z_new = np.random.multinomial(1,p_z/p_z.sum()).dot(c_ml[m,:]) 
            z[m][n]=z_new
            
            NWZ[word,z_new]+=1
            NZ[z_new]+=1
            NZM[z_new,m]+=1
        
            
            
    return z,NWZ,NZM,NZ

class Node(object):
    def __init__(self,index,parent):
        self.pop = 0
        self.children = []
        self.index=index
        self.root=False
        self.parent = parent
    
    def add_child(self, obj):
        self.children.append(obj)
        
    def decrement_pop(self):
        self.pop-=1
    
    def increment_pop(self):
        self.pop+=1
        
    def set_root(self):
        self.root=True
        
 



# code below generates initial matrix c_ml which contains topic assignments for every document m, at each stage of an L-length path l
def initialize_paths(M,K,gamma):
    tree=[]
    tree.append(Node(0,None))

    tree[0].set_root()
    next_node = 1
    c_ml = np.zeros([M,K],dtype=int)
    for m,document_vector in enumerate(c_ml):
        for l,topic in enumerate(document_vector):
            # if there are no topics selected yet, automatically select root topic as entry point to tree
            if l==0:
                c_ml[m,l]=0 # for first topic, set to root topic
                tree[0].increment_pop()
            # Otherwise, use the preceding topic's list of children to determine possible next topic/nodes
            # If there are no children in this node, then automatically fill in a new topic/node
            else:
                previous_index = c_ml[m,l-1]
            
                current_node = tree[c_ml[m,l-1]]
            
                if len(current_node.children)==0:
                    current_node.add_child(Node(next_node,current_node))
                
                    # append the most recent child to the tree list
                    tree.append(current_node.children[-1])
                    tree[-1].increment_pop()
                    c_ml[m,l] = next_node
                    next_node += 1
                
                else:
                    #print 'm,l:',m,l
                    possible_choices = tree[c_ml[m,l-1]].children # get children of parent node
                    #print 'possible_choices: ',possible_choices
                    num_possible_choices = len(possible_choices)  # each child has a chance of getting selected
                    #print 'num_possible_choice: ',num_possible_choices
                    possible_indices=np.zeros([num_possible_choices,])
                
                    possible_choice_pops=np.zeros([num_possible_choices,])
                
                    for idx,node in enumerate(possible_choices):
                        possible_indices[idx] = node.index
                        #print 'node.pop: ', node.pop
                        possible_choice_pops[idx] = node.pop
                    #print 'possible_indices: ',possible_indices   
                    possible_choice_pops = np.append(possible_choice_pops,gamma)
                    #print 'possible_choice_pops: ',possible_choice_pops
                    p = possible_choice_pops/(sum(possible_choice_pops))
                    #print 'p: ',p
                    draw = np.random.multinomial(1,p)
                    #print 'draw: ', draw
                
                    assert (draw.sum()==1.0)
                
                    if draw[-1]==1: # check to see if new table is selected
                        current_node.add_child(Node(next_node,current_node)) # add a new child to the parent
                        tree.append(current_node.children[-1])
                        tree[next_node].increment_pop()                    
                        c_ml[m,l] = next_node
                        next_node += 1 # increment index counter
                    else:
                        assert(draw[-1]==0) 
                    
                        assert len(draw) == len(possible_indices)+1
                        possible_indices=np.append(possible_indices,0)
                        new_index = draw.dot(possible_indices).astype(int) 
                    
                        assert(tree[new_index].pop != 0)
                        tree[new_index].increment_pop()
                        c_ml[m,l] = new_index
            

        

    return c_ml,tree



```
