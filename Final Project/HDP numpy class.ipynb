```python
from __future__ import division
import LDA_utils
import numpy as np
import csv
import copy
import random

     
        
        
class LDA_Gibbs_np:
    def __init__(self,K_initial,K_max,alpha,beta,V,corpus,gamma=1):
        self.M = len(corpus)
        self.alpha=alpha
        self.beta=beta
        self.V = V
        self.K = K_initial
        self.K_max = K_max
        self.U_used = range(K_initial)
        self.U_unused = range(K_initial,K_max)
        self.corpus = corpus
        self.z = copy.deepcopy(corpus)
        self.wordcount = 0
        self.M = len(corpus)
        
        self.NZW = np.zeros([K_max,V])
        self.NZ = np.zeros([K_max,])
        self.NZM= np.zeros([K_max,self.M])
        self.NM = np.zeros([self.M,])
        self.thetaZM = np.zeros([K_max,self.M])
        self.phiZW = np.zeros([K_max,V])
        self.gamma = gamma
            
        self.tau = np.ones([K_initial,])*0.1
        self.tau = np.append(self.tau,self.gamma)
               
    def initialize_assignments(self):
        print '--- setting topic assignments for each word'
        for m,document in enumerate(self.corpus):
            for n,word in enumerate(document):
                new_z_idx = random.randint(0,self.K-1)
                new_z = self.U_used[new_z_idx]
                self.NZ[new_z]+=1
                self.NZM[new_z,m]+=1
                self.NZW[new_z,word]+=1
                self.wordcount +=1
                self.NM[m]+=1
                self.z[m][n] = new_z
        
        
    def iterate_Gibbs(self,iters):
        print '--- sampling new assignments'
        print '--- NZ: ',self.NZ
        print '-- initial tau: ',self.tau
        assert np.sum(self.NZ)==np.sum(self.NM)
        assert len(self.U_used) == self.K
        for iter in range(iters):
            print 'iteration ',iter
            print '--- NZ: ',self.NZ
            print 'tau: ',self.tau
            print 'U_used: ',self.U_used
            
        
            for m,document in enumerate(self.corpus):
                for n,word in enumerate(document):
                    
                    
                    new_z = 'potato' # just to trip an error if it tries to assign new_z without redefining it.
                    
                    
                    p_z = np.zeros([self.K,])
                    self.NZ[self.z[m][n]]-=1
                    self.NZM[self.z[m][n],m]-=1
                    self.NZW[self.z[m][n],word]-=1
                    self.NM[m]-=1
                    
                    for idx,topic in enumerate(self.U_used):
                        p_z[idx]=(self.NZW[topic,word]+self.beta)/(np.sum(self.NZW[topic,:])+ self.V * self.beta)*(self.NZM[topic,m]+self.alpha*self.tau[idx])
                                   
                    
                    
                    spawn_p = self.alpha * self.tau[-1] / self.V
                    
                    
                    if self.K<self.K_max:
                        draw = random.random() * np.sum(np.append(p_z,spawn_p))
                    else:
                        draw = random.random() * np.sum(p_z)
                        
                    p = 0
                    draw_success = False
                    spawn_success = False
                    
                    for idx,k in enumerate(self.U_used[:]):
                        p +=p_z[idx]
                        assert p_z[idx]>0
                        
                        if p >= draw:
                            new_z = self.U_used[idx]
                            draw_success = True
                           
                            break
                        
                    if not draw_success:
                        new_z = self.spawn_topic()
                        
                        spawn_success = True
                        
                        print 'New topic spawned! New K value: ',self.K
                      
                        
                        
                   
                    assert draw_success or spawn_success
                    
                    
                    self.NZ[new_z]+=1
                    self.NZM[new_z,m]+=1
                    self.NZW[new_z,word]+=1
                    self.wordcount +=1
                    self.NM[m]+=1
                    self.z[m][n] = new_z
                    if spawn_success:
                        self.sample_tau()
                        print 'after spawn success, new tau: ',self.tau
                    
                    
            self.sample_tau()
            print 'after iteration, new tau: ',self.tau
            self.prune_topics()
            self.state_check()
                   
    def spawn_topic(self):
        new_topic = self.U_unused.pop(0)
        self.U_used.append(new_topic)
        print new_topic
        self.U_used.sort()
        self.U_unused.sort()
        self.K +=1
        return new_topic
    
    def delete_topic(self,topic):
        index = self.U_used.index(topic)
        print ' topic index identified for removal: ',self.U_used.index(topic)
        assert self.NZ[topic]==0
        assert np.sum(self.NZM[topic,:])==0
        assert np.sum(self.NZW[topic,:])==0
        self.U_used.remove(topic)
        self.U_unused.append(topic)
        self.U_used.sort()
        self.U_unused.sort()
        self.K -= 1
        print '--Topic about to be removed from tau. Tau befor removal: ',self.tau
        print 'U_used:',self.U_used
    
        
        self.tau = np.delete(self.tau,index) # deletes the tau entry for the index corresponding to the deleted topic
        assert len(self.tau)==len(self.U_used)+1
        print '--Topic deleted! New tau:',self.tau
        
    def prune_topics(self):
        any_deleted = False
        bool_ary = []
        for topic in self.U_used[:]:
            topic_present=False
            for document in self.z:
                for label in document:
                    if label==topic:
                        topic_present = True
                        bool_ary.append(True)
                        break
                if topic_present:
                    break
            if not topic_present:
                self.delete_topic(topic)
                any_deleted=True
        return any_deleted
            
    def sample_tau(self):
        m_k = np.zeros([self.K,])
        for t_idx,t_k in enumerate(self.U_used):
            #assert self.tau[t_idx] > 0
            count = 0
            for m in range(self.M):
                for r in range(1,(self.NZM[t_k,m]+1).astype(int)):
                    
                    if self.alpha * self.tau[t_idx] / (self.alpha*self.tau[t_idx]+r-1) > 1:
                        count += 1
                    else:                  
                        count += np.random.binomial(1,self.alpha * self.tau[t_idx] / (self.alpha*self.tau[t_idx]+r-1))
                  
            m_k[t_idx] = count
        
                
        
        alpha = np.append(m_k,self.gamma)
        
        self.tau = np.random.dirichlet(alpha)
        assert len(self.tau)==len(self.U_used)+1
    
        
                   
    def estimate_parameters(self):
        for m in range(self.M):
            for k in range(self.K_max):
                self.thetaZM[k,m] += (self.NZM[k,m]+self.alpha) / (self.NM[m]+self.K*self.alpha)
                
        for k in range(self.K_max):
            for w in range(self.V):
                self.phiZW[k,w]+=(self.NZW[k,w]+self.beta)/(self.NZ[k]+self.V*self.beta)
    def state_check(self):
        for topic in self.U_used:
            counter=0
            for document in self.z:
                for label in document:
                    if label==topic:
                        counter+=1
            assert counter == self.NZ[topic]
            assert counter == np.sum(self.NZM[topic,:])
            assert counter == np.sum(self.NZW[topic,:])
        
        
        
        
        
```
